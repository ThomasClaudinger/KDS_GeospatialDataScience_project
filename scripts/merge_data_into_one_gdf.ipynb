```python
import geopandas as gpd
import glob
import pandas as pd
```

```python
files = glob.glob('../data/*.shp')
gdf = gpd.GeoDataFrame()
for file in files:
    gdf_partial = gpd.read_file(file)
    gdf = gpd.GeoDataFrame(pd.concat([gdf, gdf_partial], ignore_index=True))
```

```python
# removing duplicates (that were created during data acquisition via API and python script)
gdf = gdf.drop_duplicates(subset='id', keep="first")
gdf
```
