```python
import pandas as pd
import json
```

```python
df = pd.read_json("../data/july_2022_raw.json")
df
```

```python
df['minutes'] = df.apply(lambda row: int(row["duration"] / 60) ,axis=1)
df
```

```python
df["minutes"].describe()
```

```python
df = df.loc[df["minutes"] <= 60]
df
```

```python
df.hist(column="minutes", bins=60)
```

```python
df = df.loc[df["minutes"] >= 5]
df
```

```python
df.hist(column="minutes", bins=55)
```

```python
df["id"] = df.index
df
```

```python
df.to_csv("july_2022_filtered.csv")
```
